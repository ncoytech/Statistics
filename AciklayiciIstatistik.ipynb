## Ames Veriseti ile Açıklayıcı İstatistik ##

Açıklayıcı İstatistik; büyük veriden anlamlı özet bilgiler getirmemizi sağlaması sebebiyle Veri Analizi'nin baş köşesindedir. 

Merkezi Eğilim : Verinin Kalbi

Merkezi Eğilim Ölçüleri
1-Aritmetik Ortalama : Değerlerin toplamının, değerlerin sayısına bölünmesiyle bulunur.      
2-Ortanca : Veriyi sıraladığımızda ortada kalan değere denir.
3-Mod : En çok tekrar eden değer veya değerlerdir.

Verinin Dağılımı
Aralık : Minimum ve maksimum değerler arasındaki farktır.
Varyans : Değerlerin ortalamadan farkının karesinin ortalamasıdır.
Standart Sapma : Varyansın kareköküdür.
Kartil Aralığı(IQR) : 25 ve 75 kartiller arasındaki aralıktır.

Verinin Boyutu ve Şekli:
Çarpıklık(Skewness) : Normal dağılımda simetri vardır, ortalama, medyan ve mod aynı noktadır. Medyan ortalamadan büyükse veri sola çarpık; tersinde ise sağa çarpıktır.


 

Sağa Çarpık ( Aritmetik ortalama > medyan > mod )



Sola çarpık ( mod > medyan > aritmetik ortalama )

Kalınlık(Kurtosis) : Dağılımın kuyruklu olma durumu; yani dağılım fazlaca çok büyük veya çok küçük aykırı değerlere sahiptir.
Yüzdelikler : Yüzdeliğe düşen değerlerdir. 25., 50, ve 75. yüzdelik dilimlere kartil denmektedir.

```python
Ames verisetindeki SalePrice özelliğine odaklanıp, istatistiki değerlerine bakalım.
```

```python
import pandas as pd
Ames = pd.read_csv(r'C:\Users\Lenovo\Desktop\portfolyo\R\Ames.csv')
 
# Descriptive Statistics of Sales Price
sales_price_description = Ames['SalePrice'].describe()
print(sales_price_description)
```

```python
median_saleprice = Ames['SalePrice'].median()
print("Median Sale Price:", median_saleprice)
 
mode_saleprice = Ames['SalePrice'].mode().values[0]
print("Mode Sale Price:", mode_saleprice)
```

Medyan değerini 159900 olarak, mod değerini 13500 olarak aldık. Bu değerleri nasıl yorumlarız? Medyan, ev fiyatlarını sıraladığımızda ortada kalan değerdir. Evlerin yarısını medyan değerin altında satılmıştır. 13500 fiyatıyla satılan ev sayısı diğer ev fiyatlarına göre daha fazla. Öyleyse ortalamayı 178.053 yüksek fyatlı evlerin etkilediğini söyleyebiliriz.

```python
range_saleprice = Ames['SalePrice'].max() - Ames['SalePrice'].min()
print("Range of Sale Price:", range_saleprice)
 
variance_saleprice = Ames['SalePrice'].var()
print("Variance of Sale Price:", variance_saleprice)
 
std_dev_saleprice = Ames['SalePrice'].std()
print("Standard Deviation of Sale Price:", std_dev_saleprice)
 
iqr_saleprice = Ames['SalePrice'].quantile(0.75) - Ames['SalePrice'].quantile(0.25)
print("IQR of Sale Price:", iqr_saleprice)
```

Ev fiyatları 12789 ile 755000 arasında  değişkenlik göstermektedir. En düşük ve en yüksek değer arasındaki fark yani aralık     742211 dir. Evlerin fiyatlarının geniş bir aralıkta değişkenlik gösterdiği söylenebilir. Evlerin %50 si 79800 civarında fiyatla satılmıştır.

```python
skewness_saleprice = Ames['SalePrice'].skew()
print("Skewness of Sale Price:", skewness_saleprice)
 
kurtosis_saleprice = Ames['SalePrice'].kurt()
print("Kurtosis of Sale Price:", kurtosis_saleprice)
 
tenth_percentile = Ames['SalePrice'].quantile(0.10)
ninetieth_percentile = Ames['SalePrice'].quantile(0.90)
print("10th Percentile:", tenth_percentile)
print("90th Percentile:", ninetieth_percentile)
 
q1_saleprice = Ames['SalePrice'].quantile(0.25)
q2_saleprice = Ames['SalePrice'].quantile(0.50)
q3_saleprice = Ames['SalePrice'].quantile(0.75)
print("Q1 (25th Percentile):", q1_saleprice)
print("Q2 (Median/50th Percentile):", q2_saleprice)
print("Q3 (75th Percentile):", q3_saleprice)
```

1.76 pozitif çarpıklık değeri, sağa doğru bir çarpıklık olduğunu ve yüksek değerli aykırı değerler bulunduğunu gösterir. 

İlk %10 luk dilimde satışların 12789 ile 107500 aralığında değişkenliği olduğunu görüyoruz, burada aykırı değer olduğunu çıkarabiliriz.
Son %10 luk dilim 272100 ve 755500 arasında dğişiyor, buradaki değişkenlik daha fazla.
Satışların %50 si 129950 ile  209750 değerleri arasında ve 159900 civarındadır.
    

```python

# Importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
 
# Setting up the style
sns.set_style("whitegrid")
 
# Calculate Mean, Median, Mode for SalePrice
mean_saleprice = Ames['SalePrice'].mean()
median_saleprice = Ames['SalePrice'].median()
mode_saleprice = Ames['SalePrice'].mode().values[0]
 
# Plotting the histogram
plt.figure(figsize=(14, 7))
sns.histplot(x=Ames['SalePrice'], bins=30, kde=True, color="skyblue")
plt.axvline(mean_saleprice, color='r', linestyle='--', label=f"Mean: ${mean_saleprice:.2f}")
plt.axvline(median_saleprice, color='g', linestyle='-', label=f"Median: ${median_saleprice:.2f}")
plt.axvline(mode_saleprice, color='b', linestyle='-.', label=f"Mode: ${mode_saleprice:.2f}")
 
# Calculating skewness and kurtosis for SalePrice
skewness_saleprice = Ames['SalePrice'].skew()
kurtosis_saleprice = Ames['SalePrice'].kurt()
 
# Annotations for skewness and kurtosis
plt.annotate('Skewness: {:.2f}\nKurtosis: {:.2f}'.format(Ames['SalePrice'].skew(), Ames['SalePrice'].kurt()),
             xy=(500000, 100), fontsize=14, bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="aliceblue"))
 
plt.title('Histogram of Ames\' Housing Prices with KDE and Reference Lines')
plt.xlabel('Housing Prices')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
